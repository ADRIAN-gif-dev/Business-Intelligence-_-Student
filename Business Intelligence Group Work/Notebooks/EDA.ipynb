```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#overview - Basic
df = pd.read_csv('/content/clean_dataset.csv')
print("Shape :", df.shape) # shape of this dataset
display(df.head())
display(df.tail())

print("\ninfo:")
df.info()

print("\nMissing values per column:")
display(df.isnull().sum())
```

```python
#some summary stats
print("\nSummary; Numerical:")
display(df.describe().T)  # transposed for ease readability

print("\nsummary; Categorical:")
display(df.describe(include=['object', 'category']).T)
```

```python
#statistics - additional
display(df.describe(include="all"))
```

```python
#analysis of correlation - if there's any numeric features
numeric_df = df.select_dtypes(include='number')

if numeric_df.empty:
    print("No numeric columns found — just skipping correlation heatmap.")
else:
    print("Plotting correlation heatmap for numeric features...")
    plt.figure(figsize=(12,6))
    sb.heatmap(numeric_df.corr(), annot=True, fmt=".2f")
    plt.show()
```

```python
# histogram for the numeric columns if any
if numeric_df.shape[1] > 0:
    numeric_df.hist(figsize=(14,10), bins=20)
    plt.suptitle("Histograms of Numeric Features")
    plt.show()
else:
    print("No numeric columns available for histograms.")
```

```python
#dup check
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# here, i was trying to value counts for categorical features
cat_cols = df.select_dtypes(include=['object', 'category']).columns

for col in cat_cols:
    print(f"\n=== Value Counts: {col} ===")
    print(df[col].value_counts())
```

```python
#data quality checkpoint
for col in df.columns:
    print(f"\nColumn: {col}")
    print("Missing:", df[col].isna().sum())
    print("Unique:", df[col].nunique())
    print("Top Value:", df[col].value_counts().idxmax())
```
